#Membaca File Gambar dan Menampilkan Gambar

```python
import cv2 
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/JohnWick.jpg', cv2.IMREAD_UNCHANGED)
cv2_imshow(img)
```

```python
print(img.shape)
```

#Menampilkan Histogram dari gambar

```python
import matplotlib.pyplot as plt

plt.hist(img.ravel(),bins = 256, range = [0,256])
plt.show()
```

#Mengubah gambar menjadi grayscale

```python
img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/JohnWick.jpg', cv2.IMREAD_GRAYSCALE)
cv2_imshow(img)
```

#Menampilkan Histogram dari gambar Grayscale
